```
puan = input("Notunuzu giriniz: ")

try:
    puan = float(puan)

    if puan >= 70:
        print("Geçti")
    else:
        print("Kaldı")

except:
    print("Lütfen sayısal bir değer giriniz.")
```

```
while True:
    puan = input("Notunuzu giriniz: ")

    try:
        puan = float(puan)

        if puan >= 90:
            print("AA")
        elif puan >= 80:
            print("BB")
        elif puan >= 70:
            print("CC")
        elif puan >= 60:
            print("DD")
        else:
            print("FF")

        break    # Doğru girişte döngüyü bitir

    except ValueError:
        print("Lütfen sayısal bir değer giriniz.")
```
